Reformat region to gene results for benchmarking

```r
library(data.table)
```

```r
path.input = "outs/region_to_gene_adj.1000000.tsv"
path.output = "pairs.E2G.res.1000000.tsv.gz"
```

```r
df.region_to_gene_adj = fread(path.input)
df.region_to_gene_adj
```

```r
df.region = do.call(rbind,strsplit(gsub(":","-",df.region_to_gene_adj$region),"-"))
df.region = as.data.frame(df.region)
colnames(df.region) = c("chr",
                        "start",
                        "end")
```

```r
df.region_to_gene_adj = cbind(df.region_to_gene_adj,
                              df.region)
```

```r
df.region_to_gene_adj[,"CellType"] = "K562"
```

```r
df.output = df.region_to_gene_adj[,c("chr",
                                     "start",
                                     "end",
                                     "target",
                                     "CellType",
                                     "importance",
                                     "rho",
                                     "importance_x_rho",
                                     "importance_x_abs_rho",
                                     "Distance"
                                     )]
```

```r
colnames(df.output) = c("chr",
                        "start",
                        "end",
                        "TargetGene",
                        "CellType",
                        "importance",
                        "rho",
                        "importance_x_rho",
                        "importance_x_abs_rho",
                        "Distance"
                        )
```

```r
df.output$sign = 0
df.output[df.output$rho >= 0,"sign"] = 1
df.output[df.output$rho < 0,"sign"] = -1
df.output$importance.signed = df.output$importance * df.output$sign
```

```r
data.table::fwrite(df.output,
                   file = path.output,
                   row.names = F,
                   quote = F,
                   sep = "\t")
df.output
```
