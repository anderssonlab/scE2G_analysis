```python
import pickle

with open('df_sparse.pkl', 'wb') as file:
    pickle.dump(df_sparse, file)
```

```python
import pickle

with open('df_sparse.pkl', 'rb') as file:
    df_sparse_loaded = pickle.load(file)
```

```python
from pycisTopic.cistopic_class import create_cistopic_object
cistopic_obj = create_cistopic_object(
    fragment_matrix = df_sparse_loaded
)
```

```python
cistopic_obj
```

```python
!mkdir -p out_model_BMMC22
```

```python
import os
os.environ['MALLET_MEMORY'] = '400G'
from pycisTopic.lda_models import run_cgs_models_mallet
# Configure path Mallet
mallet_path="Mallet-202108/bin/mallet"
# Run models
models=run_cgs_models_mallet(
    cistopic_obj,
    n_topics=[30, 40, 50],
    n_cpu=12,
    n_iter=500,
    random_state=555,
    alpha=50,
    alpha_by_topic=True,
    eta=0.1,
    eta_by_topic=False,
    tmp_path="out_model_BMMC22",
    save_path="out_model_BMMC22",
    mallet_path=mallet_path,
)
```

```python
out_dir = "outs_BMMC22"
os.makedirs(out_dir, exist_ok = True)
```

```python
import pickle
pickle.dump(
    models,
    open(os.path.join(out_dir, "models.pkl"), "wb")
)
```

```python
from pycisTopic.lda_models import evaluate_models
model = evaluate_models(
    models,
    select_model = 50,
    return_model = True
)
```

```python
cistopic_obj.add_LDA_model(model)
```

```python
pickle.dump(
    cistopic_obj,
    open(os.path.join(out_dir, "cistopic_obj.pkl"), "wb")
)
```
